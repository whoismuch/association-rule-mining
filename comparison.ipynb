```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, fpgrowth
from mlxtend.frequent_patterns import association_rules
import time
```

```python
data = pd.read_csv('Market_Basket_Optimisation.csv', header=None)
data
```

```python
processed_dataset = [[value for value in row if not pd.isnull(value)] for row in data.values.tolist()]
dataset = processed_dataset
dataset
```

```python
min_support = 20/len(dataset)
min_support
```

mlxtend
https://rasbt.github.io/mlxtend/user_guide/frequent_patterns/association_rules/

```python
te = TransactionEncoder()
te_ary = te.fit(dataset).transform(dataset)
df = pd.DataFrame(te_ary, columns=te.columns_)

start_time = time.time()
frequent_itemsets_fpgrowth = fpgrowth(df, min_support=0.03, use_colnames=True)
print(f'Ð’Ñ€ÐµÐ¼Ñ Ð²Ñ‹Ð¿Ð¾Ð»Ð½ÐµÐ½Ð¸Ñ Ð¼ÐµÑ‚Ð¾Ð´Ð°: {time.time() - start_time}')
frequent_itemsets_fpgrowth
```

```python
start_time = time.time()
frequent_itemsets_apriori = apriori(df, min_support=0.03, use_colnames=True)
print(f'Ð’Ñ€ÐµÐ¼Ñ Ð²Ñ‹Ð¿Ð¾Ð»Ð½ÐµÐ½Ð¸Ñ Ð¼ÐµÑ‚Ð¾Ð´Ð°: {time.time() - start_time}')
frequent_itemsets_apriori
```

```python
start_time = time.time()
association_rules(frequent_itemsets_fpgrowth, metric="confidence", min_threshold=0.28)
print(f'Ð’Ñ€ÐµÐ¼Ñ Ð²Ñ‹Ð¿Ð¾Ð»Ð½ÐµÐ½Ð¸Ñ Ð¼ÐµÑ‚Ð¾Ð´Ð°: {time.time() - start_time}')
```

```python
start_time = time.time()
association_rules(frequent_itemsets_apriori, metric="lift", min_threshold=1)
print(f'Ð’Ñ€ÐµÐ¼Ñ Ð²Ñ‹Ð¿Ð¾Ð»Ð½ÐµÐ½Ð¸Ñ Ð¼ÐµÑ‚Ð¾Ð´Ð°: {time.time() - start_time}')
```
